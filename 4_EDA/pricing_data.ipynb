<h3 class="numerical-vars-heading">2.1. Numerical Variables</h3>
<h4 class="variable-subheading">Price:</h4>

```python
print("### Price Analysis\n")

# Visualization: Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Product Price', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()


# Visualization: Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['price'], color='lightcoral')
plt.title('Box Plot of Product Price', fontsize=16)
plt.ylabel('Price', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()

```


  <h4>Visualization</h4>
  <ul>
    <li>
      The **Histogram** shows a **uniform distribution** of product prices,
      indicating similar frequencies for all price points from 100 to 500.
    </li>
    <li>
      The **Box Plot** confirms this, displaying the **median at 300** and
      **no visible outliers**.
    </li>
    <li>
      The **interquartile range (IQR)** for prices is from 200 to 400.
    </li>
    <li></li>
    <li></li>
  </ul>

  <h4>Insights</h4>
  <ul>
    <li>
      The analysis of **Price** reveals a **uniform distribution** across the
      100-500 range, meaning an even presence of products at every price point.
    </li>
    <li>
      The **median price is around 300**, with **no significant outliers** detected.
    </li>
    <li>
      This suggests a **consistent pricing strategy** that covers the entire
      price spectrum evenly, without focusing on specific price tiers or
      having extreme values. 📈
    </li>
  </ul>

<h3 class="numerical-vars-heading">2.3. Time-Series Variable</h3>
<h4 class="variable-subheading">Date</h4>

```python

# Convert 'date' to datetime if not already
df['date'] = pd.to_datetime(df['date'])

# Create 'month' column (start of month)
df['month'] = df['date'].dt.to_period('M').dt.to_timestamp()

# Group by 'month' and calculate mean values (ignores NaNs)
monthly_avg = df.groupby('month', as_index=False)[['price', 'discount_percent', 'competitorprice', 'finalprice']].mean()

# Sort by month just to be sure
monthly_avg = monthly_avg.sort_values('month')

# Plotting
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# Plot each line with clear style
sns.lineplot(x='month', y='price', data=monthly_avg, label='Price', marker='o')
sns.lineplot(x='month', y='discount_percent', data=monthly_avg, label='Discount %', marker='o')
sns.lineplot(x='month', y='competitorprice', data=monthly_avg, label='Competitor Price', marker='o')
sns.lineplot(x='month', y='finalprice', data=monthly_avg, label='Final Price', marker='o')

plt.title('Monthly Trends: Price, Discount %, Competitor Price & Final Price')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

<b style="font-family: 'Segoe UI', sans-serif; font-size: 17px;">🔍 Visualization:</b>
<ul style="font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.7;">
  <li><b>Line Plot:</b> Shows monthly trends of Price, Discount %, Competitor Price, and Final Price for 2024.</li>
</ul>

<b style="font-family: 'Segoe UI', sans-serif; font-size: 17px;">📌 Insights:</b>
<ul style="font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.7;">
  <li><b>Price:</b> Constant at ~300 — fixed base pricing.</li>
  <li><b>Competitor Price:</b> Stable around 315–320 — slightly higher.</li>
  <li><b>Final Price:</b> Lower at ~265–270 — reflects steady discounting.</li>
  <li><b>Discount %:</b> Just above 10 — consistent and low.</li>
  <li>📈📅 No seasonal trends — pricing strategy is stable and predictable.</li>
</ul>

<h4 class="numerical-vars-heading">3.1 Price & FinalPrice vs. Hypothetical Demand (Frequency of ProductID)</h4>


```python

# Create price bins
df['price_bin'] = pd.cut(df['price'], bins=10)
df['finalprice_bin'] = pd.cut(df['finalprice'], bins=10)

# Count of productid (proxy for demand) per price bin
price_demand = df.groupby('price_bin', observed=False)['productid'].count().reset_index().rename(columns={'productid': 'demand'})
finalprice_demand = df.groupby('finalprice_bin', observed=False)['productid'].count().reset_index().rename(columns={'productid': 'demand'})

# Plotting Price Bin vs Demand
plt.figure(figsize=(14, 6))

# Subplot 1 - Price vs Demand
plt.subplot(1, 2, 1)
sns.barplot(data=price_demand, x='price_bin', y='demand', color='skyblue')
plt.xticks(rotation=45)
plt.title('Demand vs Price')
plt.xlabel('Price Bins')
plt.ylabel('Demand (Count of ProductID)')

# Subplot 2 - FinalPrice vs Demand
plt.subplot(1, 2, 2)
sns.barplot(data=finalprice_demand, x='finalprice_bin', y='demand', color='lightgreen')
plt.xticks(rotation=45)
plt.title('Demand vs Final Price')
plt.xlabel('Final Price Bins')
plt.ylabel('Demand (Count of ProductID)')

plt.tight_layout()
plt.show()
```

<h4 class="numerical-vars-heading">3.7. Ratio Analysis: Your Price vs Competitor
</h4>


```python
# Create a new column: ratio of your price to competitor price
df['price_ratio'] = df['price'] / df['competitorprice']

# Bin the ratio and calculate demand
ratio_demand = df.groupby(pd.cut(df['price_ratio'], bins=10), observed=True)['productid'].count().reset_index()
ratio_demand.rename(columns={'productid': 'demand', 'price_ratio': 'ratio_bin'}, inplace=True)

# Bar plot for ratio
plt.figure(figsize=(10, 6))
sns.barplot(
    data=ratio_demand,
    x='ratio_bin',
    y='demand',
    hue='ratio_bin',                # Add hue to allow palette use without warning
    palette='Blues',
    edgecolor='black',
    legend=False
)

plt.xticks(rotation=45, ha='right')
plt.xlabel('Price / Competitor Price (Ratio)', fontsize=12)
plt.ylabel('Demand (Count of ProductID)', fontsize=12)
plt.title('Price Competitiveness vs Demand', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

  <h3 style="color:#2980b9;">🔻 Higher Competitiveness (Lower Ratio) = Higher Demand</h3>

    <li>This means when your product is approximately <strong>14–17% cheaper</strong> than competitors, demand is maximized.</li>
    <li>As the ratio increases (i.e., your price gets closer to or exceeds the competitor’s), demand decreases.</li>
  </ul>

  <h3 style="color:#27ae60;">⚖️ Near Parity (Ratio around 1.0) = Moderate Demand</h3>

    <li>The <strong>lowest demand (~38,000 units)</strong> occurs at a ratio of <code>(1.083, 1.111]</code>, where your product is considerably more expensive.</li>
