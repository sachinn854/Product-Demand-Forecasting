```python
print("### Price Analysis\n")

# Visualization: Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Product Price', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()


# Visualization: Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['price'], color='lightcoral')
plt.title('Box Plot of Product Price', fontsize=16)
plt.ylabel('Price', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()

```

      **no visible outliers**.
    </li>
    <li>
      The **interquartile range (IQR)** for prices is from 200 to 400.

<h3 class="numerical-vars-heading">2.3. Time-Series Variable</h3>
<h4 class="variable-subheading">Date</h4>

```python

# Convert 'date' to datetime if not already
df['date'] = pd.to_datetime(df['date'])

# Create 'month' column (start of month)
df['month'] = df['date'].dt.to_period('M').dt.to_timestamp()

# Group by 'month' and calculate mean values (ignores NaNs)
monthly_avg = df.groupby('month', as_index=False)[['price', 'discount_percent', 'competitorprice', 'finalprice']].mean()

# Sort by month just to be sure
monthly_avg = monthly_avg.sort_values('month')

# Plotting
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# Plot each line with clear style
sns.lineplot(x='month', y='price', data=monthly_avg, label='Price', marker='o')
sns.lineplot(x='month', y='discount_percent', data=monthly_avg, label='Discount %', marker='o')
sns.lineplot(x='month', y='competitorprice', data=monthly_avg, label='Competitor Price', marker='o')
sns.lineplot(x='month', y='finalprice', data=monthly_avg, label='Final Price', marker='o')

plt.title('Monthly Trends: Price, Discount %, Competitor Price & Final Price')
plt.xlabel('Month')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

<b style="font-family: 'Segoe UI', sans-serif; font-size: 17px;">ðŸ” Visualization:</b>
<ul style="font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.7;">
  <li><b>Line Plot:</b> Shows monthly trends of Price, Discount %, Competitor Price, and Final Price for 2024.</li>
</ul>

<b style="font-family: 'Segoe UI', sans-serif; font-size: 17px;">ðŸ“Œ Insights:</b>
<ul style="font-family: 'Segoe UI', sans-serif; font-size: 16px; line-height: 1.7;">
  <li><b>Price:</b> Constant at ~300 â€” fixed base pricing.</li>
  <li><b>Competitor Price:</b> Stable around 315â€“320 â€” slightly higher.</li>
  <li><b>Final Price:</b> Lower at ~265â€“270 â€” reflects steady discounting.</li>
  <li><b>Discount %:</b> Just above 10 â€” consistent and low.</li>
  <li>ðŸ“ˆðŸ“… No seasonal trends â€” pricing strategy is stable and predictable.</li>
</ul>

<h4 class="numerical-vars-heading">3.1 Price & FinalPrice vs. Hypothetical Demand (Frequency of ProductID)</h4>
